### Handling Outliers 

```python
# calculate IQR for column Assets
Q1 = df['assets'].quantile(0.25)
Q3 = df['assets'].quantile(0.75)
IQR = Q3 - Q1

# identify outliers
threshold = 1.5
outliers = df[(df['assets'] < Q1 - threshold * IQR) | (df['assets'] > Q3 + threshold * IQR)]
```

```python
outliers.shape
```

```python
outliers
```

```python
# Outliers are handled by replacing it with median values
```

```python
df.loc[outliers.index,'assets']=df['assets'].median()
df.loc[outliers.index].head(10)
```
